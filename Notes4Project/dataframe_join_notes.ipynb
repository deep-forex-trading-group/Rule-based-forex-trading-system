```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                  'data1': range(7)})

df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
    
```

```python
df1
```

```python
df2
```

### inner_join: 默认连接为inner join, 丢弃掉df2中的第二行(key==d) 和 df1中的第三行(key==c)

```python
df_default_inner_join = pd.merge(df1, df2)
df_default_inner_join
```

### outer_join: df2中没有c，df1中没有d，outer join保留两边的空值

```python
df_default_outer_join = pd.merge(df1, df2, left_on='key', right_on='key', how='outer')
df_default_outer_join
```

### left_join: df2中没有c，left join保留左边df1的所有行，右边df2留空

```python
df_default_left_join = pd.merge(df1, df2, left_on='key', right_on='key', how='left')
df_default_left_join
```

```python
df_default_left_join = pd.merge(df2, df1, left_on='key', right_on='key', how='left')
df_default_left_join
```
